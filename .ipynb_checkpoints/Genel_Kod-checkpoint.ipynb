```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model

def ShowModMedianAndDistrubition():
    readedData = pd.read_csv("data.csv")

    selectedArray = pd.Series(readedData['2P%'])

    # Calculate variance, mode, and median
    standardDev = selectedArray.values.std()

    # Display information about the DataFrame
    print(readedData.info())

    # Set the figure size before plotting
    plt.figure(figsize=(10, 6))
    # Plotting the histogram and density curve
    sns.histplot(selectedArray, kde=True, color='red', bins=8)

    # Plotting the mean line and adding text
    plt.axvline(selectedArray.mean(), linestyle='dashed', linewidth=2, color='black')
    plt.text(selectedArray.mean(), 2, 'Mean : {:.2f}'.format(selectedArray.mean()))

    # Plotting the median line and adding text
    plt.axvline(selectedArray.median(), linewidth=2, color='blue')
    plt.text(selectedArray.median(), 3, 'Median : {:.2f}'.format(selectedArray.median()))

    # Plotting the mode line and adding text
    mode_value = float(selectedArray.mode().iloc[0])  # Convert mode to float
    plt.axvline(mode_value, linestyle='dashed', linewidth=2, color='yellow')
    plt.text(mode_value, 5, 'Mode : {:.2f}'.format(mode_value), ha='right')  # Adjust the position

    # Histogram information
    plt.title('Histogram of 2-Point Field Goal Percentage / Season')
    plt.xlabel('2-Point Field Goal Percentage')
    plt.ylabel('Team Count')
    plt.xlim(selectedArray.min(), selectedArray.max())  # Adjust the range for better x-axis visualization

    # Adjusting the y-axis upper limit
    plt.ylim(0, 7.5)

    plt.show()




def ShowCorrelationHeatMap():
    plt.figure(figsize=(10, 8))  # Increase the figure size for the heatmap
    sns.heatmap(readedData.corr().round(1), annot=True, cmap='YlGnBu', fmt=".1f")
    plt.show()

def ShowPairPlot():
    sns.set()
    sns.pairplot(readedData,height = 1.5)
    plt.show()
    
def ShowLinearRegression(xColumn,yColumn):

  
    sns.regplot(x=xColumn, y=yColumn, data=readedData, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})

    plt.xlabel(xColumn)
    plt.ylabel(yColumn)

    correlation = readedData[xColumn].corr(readedData[yColumn])
    plt.text(
        readedData[xColumn].mean(),
        readedData[yColumn].mean(),
        f'Correlation: {correlation:.2f}'
    )

    plt.show()

    
readedData = pd.read_csv("data.csv")


#ShowModMedianAndDistrubition()
#ShowCorrelationHeatMap()
#ShowLinearRegression('TRB','FG%')
#ShowPairPlot()
```
