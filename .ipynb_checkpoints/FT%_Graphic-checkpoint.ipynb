```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV data
readedData = pd.read_csv("data.csv")

# Select the 'FT%' column for analysis
selectedArray = pd.Series(readedData['FT%'])

# Convert the percentage values to numeric for analysis
selectedArray_numeric = pd.to_numeric(selectedArray, errors='coerce')

# Calculate the standard deviation
standartDev = selectedArray_numeric.values.std()

# Set the figure size before plotting
plt.figure(figsize=(10, 6))  # Width: 10, Height: 6

# Plotting the histogram and density curve with more bins and specific x-axis range
sns.histplot(selectedArray_numeric, kde=True, color='red', bins=20)  # Increase the number of bins for more detail

# Plotting the mean line and adding text
plt.axvline(selectedArray_numeric.mean(), linestyle='dashed', linewidth=2, color='black')
plt.text(selectedArray_numeric.mean(), 1.5, 'Mean : {:.2f}'.format(selectedArray_numeric.mean()))

# Plotting the median line and adding text
plt.axvline(selectedArray_numeric.median(), linewidth=2, color='blue')
plt.text(selectedArray_numeric.median(), 2.5, 'Median : {:.2f}'.format(selectedArray_numeric.median()))

# Plotting the mode line and adding text
plt.axvline(selectedArray_numeric.mode().mean().astype(float), linestyle='dashed', linewidth=2, color='yellow')
plt.text(selectedArray_numeric.mode().mean(), 5, 'Mode : {:.2f}'.format(selectedArray_numeric.mode().mean()))

# Adding labels and title
plt.title('Histogram of Free Throw Percentage / Season')
plt.xlabel('Free Throw Percentage')
plt.ylabel('Team Count')

# Specify the x-axis range for better visualization
plt.xlim(selectedArray_numeric.min(), selectedArray_numeric.max())
plt.ylim(0, 5.5)  # Adjusted y-axis upper limit

# Display the plot
plt.show()
```
