```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV data
readedData = pd.read_csv("data.csv")

# Select the 'TOV' column for analysis
selectedArray = pd.Series(readedData['TOV'])

# Calculate the standard deviation
standartDev = selectedArray.values.std()
plt.figure(figsize=(10, 6))  # Adjust the width value according to your preference

# Plotting the histogram and density curve with more bins and specific x-axis range
sns.histplot(selectedArray, kde=True, color='red', bins=20)  # Increase the number of bins for more detail

# Plotting the mean line and adding text
plt.axvline(selectedArray.mean(), linestyle='dashed', linewidth=2, color='black')
plt.text(selectedArray.mean(), 1.5, 'Mean : {:.2f}'.format(selectedArray.mean()))

# Plotting the median line and adding text
plt.axvline(selectedArray.median(), linewidth=2, color='blue')
plt.text(selectedArray.median(), 2.5, 'Median : {:.2f}'.format(selectedArray.median()))

# Plotting the mode line and adding text
plt.axvline(selectedArray.mode().mean().astype(float), linestyle='dashed', linewidth=2, color='yellow')
plt.text(selectedArray.mode().mean(), 3.5, 'Mode : {:.2f}'.format(selectedArray.mode().mean()))

# Adding labels and title
plt.title('Histogram of Turnovers / Season')
plt.xlabel('Turnovers')
plt.ylabel('Team Count')

# Specify the x-axis and y-axis range for better visualization
plt.xlim(selectedArray.min(), selectedArray.max())  # Adjust the range for better x-axis visualization
plt.ylim(0, 5.5)  # Set the maximum value on the y-axis to 5

# Display the plot
plt.show()
```
